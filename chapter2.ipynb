# casting

```python
=Casting in Python means changing the data type of a variable into another type. For example, 
if you have a number stored as text (a string), and you want to treat it as a number, you can "cast" it to a different type like an integer or a float.
```

# Why use casting?


```python
=Sometimes data might be in one type, but you need to use it as another type. Casting helps to convert between types like:
From string to number (integer or float)
From number to string
From integer to float, etc
```

# How to cast in Python:

```python
int(): Converts data to an integer.
float(): Converts data to a floating-point number (with decimals).
str(): Converts data to a string (text).
```

```python
x = "10"       # x is a string
y = int(x)     # y is now an integer with the value 10
print(y)
```

```python
#String to Float:

height = "5.9"     # height is a string
height_float = float(height)  # Converts to float (5.9)
print(height_float)
```

# Unpack a Collection

```python
==Imagine you have a box (like a list or a tuple), and inside the box, there are multiple items (numbers, names, etc.).
Unpacking means taking those items out of the box and putting them into individual variables (like separate small boxes).
```

```python
fruits = ["apple", "banana", "cherry"]
# This line takes each fruit and assigns it to its own variable
fruit1, fruit2, fruit3 = fruits

# Now, each fruit is in its own variable
print(fruit1)  # This will print "apple"
print(fruit2)  # This will print "banana"
print(fruit3)  # This will print "cherry"
```

# Using the * (star) for extra items:

```python
=If you don't know how many items are in the list and only want to take the first few, you can use * to collect the rest of the items.
```

```python
numbers = [1, 2, 3, 4, 5]

# Unpack the first two numbers, and the rest will go into the 'rest' list
first, second, *rest = numbers

print(first)  # Output: 1
print(second) # Output: 2
print(rest)   # Output: [3, 4, 5]  (The remaining numbers are stored in 'rest')
```

# What is happening here?

first gets the value 1.
second gets the value 2.
rest collects the remaining values [3, 4, 5] as a list.

# Python - Global Variables

```python
=A global variable is a variable that is defined outside of a function and can be accessed by any function in the program.
Once declared, it can be used anywhere in the code, even inside functions.

=You can use a global variable inside any function or block of code.
```

```python
x = 10  # This is a global variable

def my_function():
    print(x)  # Accessing the global variable inside a function

my_function()  # Output: 10
print(x)  # Output: 10 (Accessing it outside of the function)
```

```python

# In this example, x is a global variable, so both the function and the rest of the code can use it.
```

# Modifying Global Variables inside Functions

```python
=By default, if you try to modify a global variable inside a function,
Python will treat it as a local variable (meaning it only exists within the function).
To modify a global variable inside a function, you need to use the global keyword.
```

```python
x = 10  # Global variable

def my_function():
    global x  # Telling Python to use the global 'x'
    x = 20    # Now, this changes the global 'x'

my_function()
print(x)  # Output: 20 (The global 'x' was modified inside the function)
```

# Without the global keyword, Python will think you're trying to create a new local variable inside the function, which won't affect the global x.
