# Python Strings

```python
A string in Python is a sequence of characters (like letters, numbers, symbols) 
that are written inside quotes. It's used to represent text.

you can create strings by putting text inside either single quotes (' ') or double quotes (" "). Both work the same way.
```

```python
print("Hello")
print('Hello')
```

# Looping Through a String


```python
You can use a loop to go through each letter (or character) in a string one by one. Since a string is like an "array" of characters,
you can look at each character individually using a for loop.
```

```python
for x in "banana":
    print(x)
```

# What happens here:
"banana" is the string.
The for loop takes each letter from the word "banana", one at a time.
In each loop, the letter gets printed with print(x).

# Python - Slicing Strings


```python
Slicing lets you take a part of a string by specifying where to start and where to stop.
Specify the start index and the end index, separated by a colon, to return a part of the string.
```

# Slicing

```python
Specify the start index and the end index, separated by a colon, to return a part of the string.
```

```python
In simple terms, slicing lets you take a part of a string by specifying where to start and where to stop.
```

```python
#Get the characters from position 2 to position 5 (not included):

b = "Hello, World!"
print(b[2:5])
```

```python
my_string = "Hello, World!"
print(my_string[0:5])
```

```python
[0:5] is the slice syntax.
0 is the start index (where slicing begins, starting at position 0).
5 is the end index (where slicing stops, but it doesnâ€™t include the character at position 5).
In the example above, it will return "Hello", which are the characters from position 0 to 4.
```

# Slice From the Start

```python
If you want to slice from the very beginning, you can omit the start index, and Python will automatically start from the first element (index 0).
```

```python
b = "Hello, World!"
print(b[:5])
```

```python
my_name="priyesh singh"
print(my_name[:6])
```

# Slice From the end

```python
If you want to slice to the end, you can omit the end index.
```

```python
b = "Hello, World!"
print(b[2:])
```

# Negative Indexing

```python
Use negative indexes to start the slice from the end of the string:
```

```python
b = "Hello, World!"
print(b[-5:-2])
```

```python
Slicing b[-5:-2]:

-5 refers to the 5th character from the end, which is 'r'.
-2 refers to the 2nd character from the end, which is 'd'.
In slicing, the start index (-5) is inclusive, but the end index (-2) is exclusive.
```

# Upper Case

```python
#The upper() method returns the string in upper case:

a = "Hello, World!"
print(a.upper())
```

# lower case


```python
#the lower() mwethod returns the strings in lower case
a='PREM,SINGH'
print(a.lower())
```

# Remove Whitespace

```python
Whitespace is the space before and/or after the actual text, and very often you want to remove this space.
```

```python
a = " Hello, World! "
print(a.strip()) # returns "Hello, World!"
```

# Split String

```python
The split() method returns a list where the text between the specified separator becomes the list items.
```

```python
a = "Hello, World!"
print(a.split(",")) 
```

# String Concatenation

```python
To concatenate, or combine, two strings you can use the + operator.
```

```python
a = "Hello"
b = "World"
c = a + b
print(c)
```

# String Format

```python
we can combine strings and numbers by using f-strings or the format() method!
```

```python
F-String was introduced in Python 3.6, and is now the preferred way of formatting strings.
To specify a string as an f-string, simply put an f in front of the string literal, 
and add curly brackets {} as placeholders for variables and other operations.
```

```python
age = 36
txt = f"My name is John, I am {age}"
print(txt)
```

```python
name='priyesh'
age=18
print(f"my name is {name} and age is {age}.")
```
