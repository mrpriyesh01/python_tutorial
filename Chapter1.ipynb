# Python Indentation

```python
=Indentation refers to the spaces at the beginning of a code line. In Python, indentation is not optional; it is used to define code blocks. 
Unlike other programming languages where indentation is just for readability, Python uses indentation to show the structure and flow of the program.
```

# Purpose:

```python
=Indentation in Python indicates a block of code, such as those under loops, functions, conditionals, and classes.
age = 20  # Define the variable 'age'
if age > 18:
    print("Adult")
    print("Can vote")
else:
    print("Minor")
```

# Comments in Python

```python
=Comments are notes in your code that the computer ignores. They help explain what your code does and make it easier to understand.
```

# Types of Comments:
1 Single-Line Comments:
=Start with # and go to the end of the line.
Use them for short notes.

```python
# This is a single-line comment
x = 5  # This tells that x is set to 5
```

```python
2 Multi-Line Comments:
Use multiple single-line comments or triple quotes (""" or ''') for longer notes.
```

```python
def add(a, b):
    """
    Adds two numbers and returns the result.
    """
    return a + b

# Call the function with sample inputs
result = add(3, 4)
# Print the result
print(result)  # Outputs: 7

```
